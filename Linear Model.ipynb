```python
plt.figure(figsize=(10,6))
plt.title('MaxPower vs. Price')
sns.scatterplot(data=data, x='max_power', y='selling_price');
```

```python
# price = w*Max_power+b --> (Y = m.x + c)
def estimate_charges(maxpower, w, b):
    return w * maxpower + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(data.max_power, estimate_charges(data.max_power, 50, 100), color='y', alpha=0.9);
```

```python
plt.figure(figsize=(10,6))
plt.plot(data.max_power, estimate_charges(data.max_power, 1, 10), color='y', alpha=0.9);
```

```python
plt.figure(figsize=(10,6))
plt.plot(data.max_power, estimate_charges(data.max_power, 50, 100), color='y', alpha=0.9);
plt.scatter(data.max_power, data.selling_price, s=8, alpha=0.8);
plt.xlabel('MaxPower');
plt.ylabel('Price')
plt.legend(['Estimate', 'Actual']);
```

> In order to benefit from code reuse, we will now write a function that accepts the arguments w and b.

```python
def linear_model(w, b):
    maxpower = data.max_power
    target = data.selling_price
    estimated_charges = estimate_charges(maxpower, w, b)
    plt.figure(figsize=(10,6))
    plt.plot(maxpower, estimated_charges, 'r', alpha=0.9);
    plt.scatter(maxpower, target, s=8, alpha=0.8);
    plt.xlabel('MaxPower');
    plt.ylabel('Price')
    plt.legend(['Estimate', 'Actual']);
```

```python
linear_model(0.5,1)
```

```python
linear_model(0.15,1)
```

```python
linear_model(0.15,0.6)
```

>In conclusion, we may conclude that our best fit line is obtained if w and b are both perfectly valued. The best fit line is the one that is closest to every point.
